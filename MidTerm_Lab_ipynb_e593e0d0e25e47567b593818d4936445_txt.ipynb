### Câu ví dụ: Đếm bội số

### Có bao nhiêu bội số của 7 trong các số nguyên từ 1 đến 1001?

#### Lời giải ví dụ:

##### 1. Code đếm lời giải (Vòng lặp có điều kiện)

```python
N = 1001
divisor = 7
count_loop = 0
for i in range(1, N + 1):
    if i % divisor == 0:
        count_loop += 1
print(f"Code đếm bằng vòng lặp: {count_loop}")
```

##### 2. Lời giải bằng Toán, có giải thích

Đây là bài toán đếm số phần tử trong một dãy số chia hết cho k

- Công thức: Số bội số của k trong khoảng $[1,N]$ là $\lfloor \frac{n}{k} \rfloor$.
- Giải thích: Ta cần tìm số nguyên sao $m$ cho $7m <= 1001$

$$ m = \lfloor \frac{1001}{7} \rfloor = 143$$

##### 3. Code so sánh lời giải

```python
import math
N = 1001
divisor = 7
count_math = math.floor(N / divisor)
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```

### Câu 1:

### Có bao nhiêu số chẵn từ 100 đến 999 có các chữ số đôi một khác nhau?

## Giải bằng toán học
####**+ Gọi "ABC" là số từ 100 đến 999**
 - Cách chọn A : 1 -> 9
 - Cách chọn B : 1 -> 9
 - Cách chọn C : 0, 2, 4, 6, 8

####**+ TH1: C = 0**
 - C : 1 cách chọn
 - B : 9 cách chọn
- A : 8 cách chọn
 #### --> Có 1 x 9 x 8 = 72 cách
#### **+ TH2: C != 0**
- C : 4 cách chọn
 - A : 8 cách chọn
 - B : 8 cách chọn
 --> có 4 x 8 x 8 = 256 cách
#### **==> Vậy có tổng cộng 256 + 72 = 328 cách chọn**

```python
count_loop = 0
for i in range(100, 999):
  if i % 2 == 0:
    s = str(i)
    if(s[0] != s[1] and s[1] != s[2] and s[0] != s[2]):
      count_loop = count_loop + 1
```

```python
count_math = 1*9*8 + 4*8*8
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```

### Bao nhiêu số lẻ 5 chữ số có chính xác một cặp chữ số trùng (tất cả chữ số khác đều khác nhau), và chữ số cuối là lẻ?

## **Giải bằng toán học**
#### Gọi số cần tìm là abcde
#### Ràng buộc:
- e có 5 cách chọn: 1, 3, 5, 7, 9
- a có 9 cách chọn: 1 -> 9
####Xét: Vị trí trùng nằm ở e
TH1: a = e  
 - a và e có 5 cách chọn (thỏa a != 0)  
 - 3 vị trí còn lại có 9P3 cách chọn  
####-> Có 5 * 9P3 = 2520 (Cách)  
TH2: a != e  
 - e và 1 vị trí i nào đó có 5 cách chọn   
 - a có 8 cách chọn
 - 2 vị trí còn lại có 8P2 cách chọn
####-> Có 3 * 5 * 8 * 8P2 = 6720 (Cách)  
### Vậy vị trí trùng ở e có tổng cộng 2520 + 6720 = 9240 (Cách)
####Xét: Vị trí trùng không nằm ở e
TH1: nằm ở a
- e có 5 cách chọn
- a và vị trí trùng còn lại có 8 cách chọn
- 2 vị trí còn lại có 8P2 cách chọn
####-> Có 3 * 5 * 8 * 8P2 = 6720 (Cách)  
TH2: không nằm ở a  
 TH2.1: giá trị trùng != 0  
  - e có 5 cách chọn
  - cặp trùng có 8 cách chọn
  - a có 7 cách chọn
  - vị trí còn lại có 7 cách chọn  
TH2.2: giá trị trùng = 0  
  - e có 5 cách chọn
  - cặp trùng 1 cách chọn
  - a có 8 cách chọn
  - vị trí còn lại có 7 cách chọn
####-> 3 * 5 * 8 * 7 * 7 + 3 * 5 * 1 * 8 * 7 = 6720 (Cách)  
### Vậy vị trí trùng không nằm ở e có tổng cộng 6720 + 6720 = 13440 (Cách)

## Tổng: 9240 + 13440 = 22680 (Cách)



```python
count_loop = 0
for i in range(10000, 100000):
    if i % 2 == 0:
        continue
    s = str(i)
    valid = True
    pair = 0
    for j in range(len(s)):
        freq = s.count(s[j])
        if freq > 2:
            valid = False
            break
        if freq == 2:
             pair += 1
    if valid and pair == 2:
        count_loop += 1
```

```python
count_math = 9240 + 13440
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```

### Câu 3:
### Từ các ký tự $ {a,b,c,d,e,f} $, có bao nhiêu cách tạo ra một chuỗi gồm 3 ký tự tùy ý?

## Giải bằng toán học
### Có 3 vị trí, mỗi vị trí có 6 cách chọn ký tự
###-> 6^3 cách

```python
from itertools import permutations
chars = ['a', 'b', 'c', 'd', 'e', 'f']
count_loop = 0
for a in chars:
    for b in chars:
        for c in chars:
            count_loop += 1
```

```python
count_math = pow(6, 3)
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```

## Giải bằng toán học
### Chọn 3 ký tự từ 6 ký tự, có thể hoán đổi -> dùng chỉnh hợp

### Câu 8:

### Từ một nhóm gồm 8 nhà khoa học nam và 6 nhà khoa học nữ, có bao nhiêu cách thành lập một ủy ban gồm 6 người, sao cho ủy ban có ít nhất 3 nhà khoa học nữ?

## Giải bằng toán học
### Ta chia thành các trường hợp theo số lượng nữ trong ủy ban:
### TH1: Có 3 nữ -> 6C3 * 8C3
### TH2: Có 4 nữ -> 6C4 * 8C2
### TH3: Có 5 nữ -> 6C5 * 8C1
### TH4: Có 6 nữ -> 6C6 * 8C0
### -> Cộng tất cả các trường hợp lại được 1589 cách



```python
from itertools import combinations
count_loop = 0
scientists = ['M'] * 8 + ['F'] * 6
for combo in combinations(scientists, 6):
  if combo.count('F') >= 3:
    count_loop += 1
```

```python
from math import comb
count_math = comb(6, 3) * comb(8, 3) + comb(6, 4) * comb(8, 2) + comb(6, 5) * comb(8, 1) + comb(6, 6) * comb(8, 0)
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```
