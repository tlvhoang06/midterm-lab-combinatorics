### Câu ví dụ: Đếm bội số

### Có bao nhiêu bội số của 7 trong các số nguyên từ 1 đến 1001?

#### Lời giải ví dụ:

##### 1. Code đếm lời giải (Vòng lặp có điều kiện)

```python
N = 1001
divisor = 7
count_loop = 0
for i in range(1, N + 1):
    if i % divisor == 0:
        count_loop += 1
print(f"Code đếm bằng vòng lặp: {count_loop}")
```

##### 2. Lời giải bằng Toán, có giải thích

Đây là bài toán đếm số phần tử trong một dãy số chia hết cho k

- Công thức: Số bội số của k trong khoảng $[1,N]$ là $\lfloor \frac{n}{k} \rfloor$.
- Giải thích: Ta cần tìm số nguyên sao $m$ cho $7m <= 1001$

$$ m = \lfloor \frac{1001}{7} \rfloor = 143$$

##### 3. Code so sánh lời giải

```python
import math
N = 1001
divisor = 7
count_math = math.floor(N / divisor)
print(f"Code đếm bằng vòng lặp: {count_loop}")
print(f"Code sử dụng công thức: {count_math}")
print(f"Kết quả so sánh: {count_loop == count_math}")
```

### Có bao nhiêu số chẵn từ 100 đến 999 có các chữ số đôi một khác nhau?

## **+ Gọi "ABC" là số từ 100 đến 999**
### - Cách chọn A : 1 -> 9
### - Cách chọn B : 1 -> 9
### - Cách chọn C : 0, 2, 4, 6, 8

## **+ TH1: C = 0**
 ### C : 1 cách chọn
 ### B : 9 cách chọn
 ### A : 8 cách chọn
 ### --> có 1 x 9 x 8 = 72 cách
## **+ TH2: C != 0**
### C : 4 cách chọn
### A : 8 cách chọn
### B : 8 cách chọn
### --> có 4 x 8 x 8 = 256 cách
## **==> Vậy có tổng cộng 256 + 72 = 328 cách chọn**

```python
count = 0
for i in range(100, 999):
  if i % 2 == 0:
    s = str(i)
    if(s[0] != s[1] and s[1] != s[2] and s[0] != s[2]):
      count = count + 1
```

### Bao nhiêu số lẻ 5 chữ số có chính xác một cặp chữ số trùng (tất cả chữ số khác đều khác nhau), và chữ số cuối là lẻ?

# **Giải bằng toán học**
## + Gọi số cần tìm là abcde
## + Ràng buộc:
## - e có 5 cách chọn: 1, 3, 5, 7, 9
## - a có 9 cách chọn: 1 -> 9
## + Xét: Vị trí trùng nằm ở e
## TH1: a = e
## a và e có 5 cách chọn (thỏa a != 0)
## 3 vị trí còn lại có 9P3 cách chọn
## -> Có 5 * 9P3 = 2520 (Cách)
## TH2:



```python
count = 0
for i in range(10000, 100000):
    if i % 2 == 0:
        continue
    s = str(i)
    valid = True
    pair = 0
    for j in range(len(s)):
        freq = s.count(s[j])
        if freq > 2:
            valid = False
            break
        if freq == 2:
             pair += 1
    if valid and pair == 2:
        count += 1
print(count)
```
